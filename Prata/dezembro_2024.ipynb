```python
df_dezembro = pd.read_excel(r'C:\Users\JOSEP\Documents\Projeto--An-lise-da-Desigualdade-de-Renda-e-Emprego-no-Brasil\Bronze\12-tabelas_dezembro de 2024 - Site.xlsx',sheet_name=7)
df_dezembro
```

```python
df_dezembro.drop(df_dezembro.columns[0], axis=1, inplace=True)
```

```python
df_dezembro = df_dezembro.dropna(how='any')

df_dezembro.columns = df_dezembro.iloc[0]

df_dezembro = df_dezembro[1:].reset_index(drop=True)

df_dezembro.columns = ['Mês', 'Estoque', 'Admissões', 'Desligamentos', 'Saldos', 'Variação Relativa (%)']

df_dezembro
```

```python
meses = {
    'janeiro': '01',
    'fevereiro': '02',
    'março': '03',
    'abril': '04',
    'maio': '05',
    'junho': '06',
    'julho': '07',
    'agosto': '08',
    'setembro': '09',
    'outubro': '10',
    'novembro': '11',
    'dezembro': '12'
}

df_dezembro['Mês'] = df_dezembro['Mês'].astype(str).str.lower()

df_dezembro['Data_str'] = df_dezembro['Mês'].apply(lambda x: meses[x.split('/')[0]] + '/01/' + x.split('/')[1])

df_dezembro['Mês'] = pd.to_datetime(df_dezembro['Data_str'], format='%m/%d/%Y')

df_dezembro.drop(columns=['Data_str'], inplace=True)

df_dezembro
```

```python
df_dezembro.to_excel(r'C:\Users\JOSEP\Documents\Projeto--An-lise-da-Desigualdade-de-Renda-e-Emprego-no-Brasil\Ouro\Dados_Transformados.xlsx', index=False)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(16, 14))

# Gráfico 1: Estoque
sns.lineplot(data=df_dezembro, x='Mês', y='Estoque', ax=axs[0, 0], marker='o')
axs[0, 0].set_title('Estoque ao longo do tempo')
axs[0, 0].tick_params(axis='x', rotation=45)

# Gráfico 2: Admissões
sns.lineplot(data=df_dezembro, x='Mês', y='Admissões', ax=axs[0, 1], marker='o')
axs[0, 1].set_title('Admissões ao longo do tempo')
axs[0, 1].tick_params(axis='x', rotation=45)

# Gráfico 3: Desligamentos
sns.lineplot(data=df_dezembro, x='Mês', y='Desligamentos', ax=axs[1, 0], marker='o')
axs[1, 0].set_title('Desligamentos ao longo do tempo')
axs[1, 0].tick_params(axis='x', rotation=45)

# Gráfico 4: Saldos
sns.lineplot(data=df_dezembro, x='Mês', y='Saldos', ax=axs[1, 1], marker='o')
axs[1, 1].set_title('Saldos ao longo do tempo')
axs[1, 1].tick_params(axis='x', rotation=45)

# Gráfico 5: Variação Relativa (%)
sns.lineplot(data=df_dezembro, x='Mês', y='Variação Relativa (%)', ax=axs[2, 0], marker='o')
axs[2, 0].set_title('Variação Relativa (%) ao longo do tempo')
axs[2, 0].tick_params(axis='x', rotation=45)

# Gráfico 6: Comparação Admissões vs Desligamentos
sns.lineplot(data=df_dezembro, x='Mês', y='Admissões', ax=axs[2, 1], label='Admissões', marker='o')
sns.lineplot(data=df_dezembro, x='Mês', y='Desligamentos', ax=axs[2, 1], label='Desligamentos', marker='o')
axs[2, 1].set_title('Admissões vs Desligamentos')
axs[2, 1].tick_params(axis='x', rotation=45)
axs[2, 1].legend()

# Ajusta os espaços
plt.tight_layout()
plt.show()
```
